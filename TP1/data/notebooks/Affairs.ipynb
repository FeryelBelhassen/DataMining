````python
affairs_df = pd.read_csv("../data/sources/Affairs.txt")
````

````python
temp_df = affairs_df.append(affairs_df)

temp_df.shape
````

````python
temp_df.drop_duplicates(inplace=True)
````

````python
temp_df = affairs_df.append(affairs_df)  # make a new copy

temp_df.drop_duplicates(inplace=True, keep=False)

temp_df.shape
````

````python
affairs_df.columns
````

````python
affairs_df.rename(columns={
        'affairs': 'Affairs', 
        'gender': 'Gender',
        'age': 'Age',
        'yearsmarried': 'Years_married',
        'children': 'Children',
        'religiousness': 'Religiousness',
         'education':  'Education',
        'occupation': 'Occupation',
        'rating': 'Rating'
    }, inplace=True)


affairs_df.columns
````

````python
affairs_df.columns = ['Unamed', 'affairs', 'gender', 'age', 'yearsmarried', 'children', 'religiousness', 
                     'education', 'occupation', 'rating']


affairs_df.columns
````

````python
affairs_df.columns = [col.lower() for col in affairs_df]

affairs_df.columns
````
